```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randint(1,100,(4,3)),index=['Fenil','Teerth','Naitik','Aastha'],columns=['Python','DataScience','DBMS'])
df
```

```python
plt.plot(df.index,df.Python,'b-.')
```

```python
plt.plot(df.index,df.Python,'y-.',label='Python',lw='0.25',marker='1')
plt.plot(df.index,df.DBMS,'r-.',label='DBMS')
plt.plot(df.index,df.DataScience,'k-.',label='DataScience')
plt.title('Python marks')
plt.xlabel('names')
plt.ylabel('marks')
plt.legend()
plt.legend(loc=4)#for changing side of legend
plt.ylim(-20,100)
```

```python
plt.subplot(2,2,1)
plt.plot(df.index,df.Python,'y-.',label='Python',lw='0.25',marker='1')
plt.subplot(2,2,2)
plt.plot(df.index,df.DBMS,'r-.',label='DBMS')
plt.subplot(2,2,3)
plt.plot(df.index,df.DataScience,'k-.',label='DataScience')
plt.title('Python marks')
plt.xlabel('names')
plt.ylabel('marks')
plt.legend()
plt.legend(loc=4)#forchanging side of legend
plt.ylim(-20,100)

plt.subplot(2,2,2)
plt.plot(df.index,df.DBMS,'r-.',label='DBMS')
plt.subplot(2,2,3)
plt.plot(df.index,df.DataScience,'k-.',label='DataScience')
plt.title('Python marks')
plt.xlabel('names')
plt.ylabel('marks')
plt.legend()
plt.legend(loc=4)#forchanging side of legend
plt.ylim(-20,100)

plt.subplot(2,2,4)
plt.plot(df.index,df.Python,'y-.',label='Python',lw='0.25',marker='1')
plt.plot(df.index,df.DataScience,'k-.',label='DataScience')
plt.tight_layout()
```
